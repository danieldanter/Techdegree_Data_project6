```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv

sns.set_theme()

video_games = pd.read_csv('top100videogamesales.csv')

video_games.head(50)
```

# 1: Regional Sales by Publisher: 

```python
# format the dataframe -> so we can use it for the stacked barchart
Sales_by_pub = video_games.drop(['Rank', 'Name','Platform','Year','Genre','Global_Sales'], axis=1)
Sales_by_pub= Sales_by_pub.groupby(['Publisher'])[['NA_Sales','EU_Sales', 'JP_Sales','Other_Sales']].sum()
Sales_by_pub.head(15)
```

```python
# create stacked bar chart for Regional Sales by Publisher
Sales_by_pub.plot(kind='bar', stacked=True)
 
# labels for x & y axis
plt.xlabel('Publisher')
plt.ylabel('Total Sales in Dollers')
 
# title of plot
plt.title('Regional Sales by Publisher')
```

## This chart shows the Regional Sales by Publisher -> Nintento has the most sales

## 2: Global Sales by Year: 

```python
# format the dataframe -> so we can use it for the line chart
#Sales_by_Year = video_games.drop(['Rank', 'Name','Platform','Publisher','Genre','NA_Sales','EU_Sales', 'JP_Sales','Other_Sales'], axis=1)
Sales_by_Year= video_games.groupby(['Year'])[['Global_Sales']].sum()
Sales_by_Year.head(50)
```

```python
sns.lineplot(data=Sales_by_Year, x= 'Year', y='Global_Sales')
# labels for x & y axis
plt.xlabel('Year')
plt.ylabel('Total Sales in Dollers')
 
# title of plot
plt.title('Global Sales by Year')
```

## This chart shows the Global Sales over Time from 2001 to 2015

# 3:Global Sales Distribution for Action Games: 

```python
# get only the action games
action = video_games.loc[video_games['Genre'] == 'Action']
#plot the action games
sns.histplot(data= action, x='Global_Sales', bins=10, kde=True)
# labels for x & y axis
plt.ylabel('Nr of Games')
plt.xlabel('Globals Sales')
 
# title of plot
plt.title('Global Sales Distribution for Action Games')
```

## there are 5 Games with Globel Sales in the range 7-8 and  one in the range 8-10 . two in the range 10-11 and two in the range 11-12 ......... and two in the range 20-21

# 4: North American Sales by Genre: 

```python
# number of top-n you want
n = 7
# group by & sort descending
df_sorted = (video_games
                .groupby('Genre').sum()
                .sort_values('NA_Sales', ascending=False)
                .reset_index()
            )

#df_sorted .head(20)

# rename rows other than top-n to 'Others'
df_sorted.loc[df_sorted.index >= n, 'Genre'] = 'Others (Simulation, Fighting, Puzzle)'

# re-group by again
with_others = df_sorted.groupby('Genre').sum()

with_others.head(50)
```

```python
#draw plot
with_others.plot(kind='pie', y='NA_Sales',autopct='%1.0f%%')

# labels for x & y axis
plt.ylabel('')
# title of plot
plt.title(' North American Sales by Genre')
# legend of plot
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper left', borderaxespad=0)
```

## Most NA Sales very made in the category shooter. Second place Sports and third Place is Action

# European Sales by Genre:

```python
# number of top-n you want
n = 7
# group by & sort descending
df_sorted = (video_games
                .groupby('Genre').sum()
                .sort_values('EU_Sales', ascending=False)
                .reset_index()
            )

df_sorted .head(20)

# rename rows other than top-n to 'Others'
df_sorted.loc[df_sorted.index >= n, 'Genre'] = 'Others (Simulation, Fighting, Puzzle)'

# re-group by again
with_others = df_sorted.groupby('Genre').sum()

with_others.head(50)
```

```python
#draw plot
with_others.plot(kind='pie', y='EU_Sales',autopct='%1.0f%%')

# labels for x & y axis
plt.ylabel('')
# title of plot
plt.title(' European Sales by Genre')
# legend of plot
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper left', borderaxespad=0)
```

## Most European Sales very made in the category shooter. Second place(not by much) Sports and third Place is Action
